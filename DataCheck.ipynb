```python
###SYNTHETIC DATA
import json
import os
import shutil
#Empty check.
problem_map = json.load(open('./datasets/benchmarks/unibo-problems.json'))
duplicate_check = {}
main_directory="./results_synth_data/"
for foldername in os.listdir(main_directory):
    folder = os.path.join(main_directory, foldername)
    if os.path.isdir(folder):
        params = False
        metrics = False
        dataset = False
        for filename in os.listdir(folder):
            if filename.find("parameters") != -1:
                params=True
            elif filename.find("metrics") != -1:
                metrics=True
            elif filename.find("dataset") != -1:
                dataset=True
        if not params or not metrics or not dataset:
            print(folder)
```

```python
#Count problem seeds.
import json
import pandas as pd
import pickle
from analysis_util import *
parse_synthetic_results(problem_map, save_dict=True)
filehandler = open('parsed_results/synth_results.obj', 'rb') 
synth_results = pd.DataFrame.from_dict(pickle.load(filehandler))
```

```python
#Duplicate check.
duplicate_counter = {}
for index, row in synth_results.iterrows():
    key = row['surrogate'] + "_" + row['acquisition'] + "_"+ str(row['dim']) + "_" + str(row['problem_idx']) + "_" + str(row['seed'])
    if key not in duplicate_counter.keys():
        duplicate_counter[key] = row['path']
    else:
        print(duplicate_counter[key])
        print(row['path'])
```

```python
#SEED COUNT
seed_counter = {}
expected_seeds = 20
for index, row in synth_results.iterrows():
    key = row['surrogate'] + "_" + row['acquisition'] + "_"+ str(row['dim']) + "_" + str(row['problem_idx'])
    if key not in seed_counter.keys():
        seed_counter[key] = 1
    else:
        seed_counter[key] +=1
for key in seed_counter.keys():
    if seed_counter[key] != expected_seeds:
        print(key)
```
