```python
### GENERATE RECALIBRATION AND VANILLA DIFF TABLES
filehandler = open('parsed_results/real_results.obj', 'rb') 
real_results_vanilla = pd.DataFrame.from_dict(pickle.load(filehandler))
filehandler = open('parsed_results/real_results_recal.obj', 'rb') 
real_results_recal = pd.DataFrame.from_dict(pickle.load(filehandler))
```

```python
real_results_vanilla = real_results_vanilla.drop(real_results_vanilla.index[(real_results_vanilla['surrogate']=="BNN") & (real_results_vanilla['seed']==11) & (real_results_vanilla['data']=="mnist_cnn") & (real_results_vanilla['acquisition']=="RS")].tolist())
real_results_vanilla = real_results_vanilla.sort_values(by=['surrogate', 'acquisition', 'data', 'seed'])
real_results_recal = real_results_recal.sort_values(by=['surrogate', 'acquisition', 'data', 'seed'])
aggregate_real_results_vanilla = aggregate_real_results(real_results_recal)
aggregate_real_results_recal = aggregate_real_results(real_results_vanilla)
```

```python
real_results_diff = aggregate_real_results_vanilla.copy()
columns = ['tot_regret_pool', 'inst_regret_pool', 'calibration_mse', 'sharpness', 'elpd']
for column in columns:
    real_results_diff[column] = aggregate_real_results_vanilla[column]-aggregate_real_results_recal[column]
```

```python
real_results_highlighted, latex_results = BO_performance_table1_real_results(real_results_diff)
#display(aggregate_real_results_df)
display(real_results_highlighted)
display(latex_results)
with open("./tables/real_performance_recal_diff.tex", "w") as file:
    latex_str = latex_results.to_latex(index=False)
    latex_str = latex_str.replace('\\textbackslash pm', '$\pm$')
    print(latex_str)
    file.write(latex_str)
###DIFF RECALIBRATE REAL DATA DONE
```
